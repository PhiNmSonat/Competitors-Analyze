```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a pandas DataFrame
df = pd.read_csv('D:\\Downloads\\VSCode\\Python\\DA\\onjobtrain\\screwjamreviews.csv')

# Step 2: Convert the 'review_date' column to datetime format
df['review_date'] = pd.to_datetime(df['review_date'])

# Step 3: Group the data by 'review_date' and 'rating' and count the occurrences
rating_counts = df.groupby([df['review_date'].dt.date, 'rating']).size().unstack(fill_value=0)

# Step 4: Plot the line chart
plt.figure(figsize=(20, 5))

# Plot each rating count as a separate line with different colors
for rating in range(1, 6):
    plt.plot(rating_counts.index, rating_counts[rating], label=f'{rating} star')

plt.title('Rating Count Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```
